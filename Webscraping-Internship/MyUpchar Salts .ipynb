```python
# given are the salt details which are going to be searched and further their medicines details will be scraped 
import pandas as pd
mediFee=pd.read_csv('MediFEE_salts.csv')
l=mediFee['Salt Name']
saltList=l.head(42)
saltList
```

```python
driver=webdriver.Chrome(ChromeDriverManager().install())
driver.maximize_window()
driver.get('https://www.myupchar.com/en/medicine')
driver.implicitly_wait(10)

def list_links(res):
    
    l=[]
    for tag in res:
        l.append(tag.get_attribute('href'))
    return l    
    
def get_med_links(salt):
    driver.find_element(By.XPATH,"//a[contains(@id,'search')]").click()
    driver.find_element(By.XPATH,"//input[contains(@id,'searchInputTop')]").send_keys(salt)
    driver.find_element(By.XPATH,"//input[contains(@id,'searchInputTop')]").submit()
    driver.implicitly_wait(10)
    driver.find_element(By.XPATH,"//a[contains(@data-value,'medicine')]").click()
    #med_names=driver.find_elements(By.XPATH,"//h5") 
    driver.implicitly_wait(10)
    med_links=driver.find_elements(By.XPATH,"//a[contains(@class,'surgery-section-bx')]")
    
    return list_links(med_links)

saltLinks=dict()
for salt in saltList[1:]:
    saltLinks[salt]=get_med_links(salt)
saltLinks   
```

```python
from bs4 import BeautifulSoup
data=[]
driver=webdriver.Chrome(ChromeDriverManager().install())
driver.maximize_window()

def element(warning):
    l=warning.get_attribute('outerHTML')
    doc= BeautifulSoup(l, 'html.parser') 
    x=doc.find_all('p')
    y=[i.text for i in x]
    return y[1::2]
        
def get_med_details(salt,medicine,prescription,price,manufacturer,related_warnings):
    try:
        l=[]
        l.append(salt) #salt
        l.append(medicine.text) #med name
        l.append(price.text)
        y=(manufacturer.text).split('\n')
        l.append(y[0].split(':')[1])  #manufacturer
        try:
            p=prescription.text
        except:
            p=prescription
        if p == 'Prescription Required':
            l.append("Yes")
        else:
            l.append("No")

        result=element(related_warnings[0])
        res=element(related_warnings[2])

        for i in res:
            l.append(i)
        for i in result:
            l.append(i)
    except:
        l.append('NaN')
    data.append(l)

def med_details_webelement(salt,i):
    global data
    try:
        driver.get(i)
        medicine=driver.find_elements(By.XPATH,"//h1")[1]
        prescription=driver.find_element(By.XPATH,"//span[contains(@class,'pres_txt')]")
        price=driver.find_element(By.XPATH,"//span[contains(@class,'txt_big')]")

        manufacturer=driver.find_element(By.XPATH,"//ul[contains(@class,'medicine_info')]") 
        related_warnings=driver.find_elements(By.XPATH,"//ul[contains(@class,'med_items')]") #will contains 3 elements
        get_med_details(salt,medicine,prescription,price,manufacturer,related_warnings)
        
    
    except:
        prescription='No prescription'
    
       
for salt,links in saltLinks.items():
    for i in links:
        med_details_webelement(salt,i)
        
        
```
