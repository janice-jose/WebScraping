## Using beautiful soup to parse and extract information

```python
!pip install beautifulSoup4
```

```python
from bs4 import BeautifulSoup
doc= BeautifulSoup(page_contents, 'html.parser')
```

```python
#extracting popular topics
class_name='f3 lh-condensed mb-0 mt-1 Link--primary'
topic_title=doc.find_all('p',{'class':class_name})
print(len(topic_title))
topic_title[:5]
```

```python
desc_class="f5 color-text-secondary mb-0 mt-1"             
topic_desc=doc.find_all('p',{'class':desc_class})
topic_desc[:5]
```

```python
topic_link_class="d-flex no-underline" 
topic_link=doc.find_all('a',{'class':topic_link_class})
print(topic_link[1]['href'])
topicUrl="https://github.com"+topic_link[1]['href']
topicUrl
```

```python
topics_title=[]
for tag in topic_title:
    topics_title.append(tag.text)
print(topics_title[:3])    

topic_descs=[]
for tag in topic_desc:
    topic_descs.append(tag.text.strip())
print(topic_descs[:3])    

Url_topic=[]
base_url="https://github.com"
for tag in topic_link:
    Url_topic.append(base_url+tag['href'])
Url_topic[:3]   
```

```python
 import pandas as pd
```

```python
dict={'title':topics_title,'description':topic_descs,'URL':Url_topic}
topics_df=pd.DataFrame(dict)
topics_df
```

## Create Csv from extracted information

```python
topics_df.to_csv('topics.csv')
```

```python
star_class="social-count float-none"
star_link=doc.find_all('a',{'class':star_class})
star_link[0].text.strip()
def remove_k(string):
    if string[-1]=='k':
        no=int(float(string[:-1])*1000)
    else:
        no=int(string)
    return no
remove_k(star_link[0].text.strip())
           
```

```python
#function for getting popular repo name, link ,username ,star of a single topic

def get_repo_details(h_link,star_link,title):
    user_name=[]
    repo_name=[]
    repo_link=[]
    star_total=[]
    topic=title
    for i in h_link:
        a_tags=i.find_all('a')
        user_name.append(a_tags[0].text.strip())
        repo_name.append(a_tags[1].text.strip())
        repo_link.append('https://github.com'+a_tags[1]['href'])
    for i in star_link:
        star_link=doc.find_all('a',{'class':star_class})
        string=i.text.strip()
        if string[-1]=='k':
            star_total.append(int(float(string[:-1])*1000))
        else:
            star_total.append(int(string))
    dict={'Topic':topic,'Username':user_name,'Repo_name': repo_name,'URL':repo_link,'Stars':star_total};
    popular_repo=pd.DataFrame(dict);
    return(popular_repo);
            

def link_tags(topic_url):
    topic_page_content=topic_url.text
    
    class_name="f3 color-text-secondary text-normal lh-condensed"
    star_class="social-count float-none"
    
    links=[]
    doc= BeautifulSoup(topic_page_content, 'html.parser')
    links.append(doc.find_all('h3',{'class':class_name}))
    links.append(doc.find_all('a',{'class':star_class}))
    
    return links;

#function to loop through all popular repos 
def get_topic_repos(URL,title):

    response=requests.get(URL)

    if response.status_code!=200:

        raise Exception('Link for '+ title + 'not found')
    else:
        x=link_tags(response)
        h_link=x[0]
        star_link=x[1]
        return get_repo_details(h_link,star_link,title);
        
```

```python
get_topic_repos(topics_df['URL'][0],topics_df['title'][0])
```

```python
#for each topic 
frames=[]
for i in range(len(topics_df)):
    frames.append(get_topic_repos(topics_df['URL'][i],topics_df['title'][i]));
    
result=pd.concat(frames)    
    
```

```python
result.to_csv('popular_repos.csv')
```
